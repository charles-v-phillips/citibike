```python
from bs4 import BeautifulSoup
import requests
import zipfile
import os
import wget

url = 'https://s3.amazonaws.com/tripdata/'
request = requests.get(url)
soup = BeautifulSoup(request.text, 'xml')

#Find all links on page
links = soup.find_all('Key')
links = [links[i].get_text() for i in range(len(links)-1)]
for i ,link in enumerate(links):
    print(i)
    wget.download(url + link)
    zip_ref = zipfile.ZipFile(link,allowZip64 = True)
    zip_ref.extractall('.')
    zip_ref.close()
    os.remove(link)





```
