```python
# assigning season

def get_season(row):
    """
    convert date to month and day as integer (md), e.g. 4/21 = 421, 11/17 = 1117, etc.
    """
    m = row['started_at'].month * 100
    d = row['started_at'].day
    md = m + d

    if ((md >= 320) and (md <= 621)):
        s = 'spring' # spring
    elif ((md > 621) and (md < 923)):
        s = 'summer'  # summer
    elif ((md >= 923) and (md <= 1220)):
        s = 'autumm'  # fall
    elif ((md > 1220) or (md < 320)):
        s = 'winter'  # winter
    else:
        raise IndexError("Invalid date")

    return s
```

```python
year2019['season'] = year2019.apply(lambda r: get_season(r), axis = 1)
```

```python
# sample of getting distance 

import haversine as hs
from haversine import Unit
loc1 = (28.426846, 77.088834)
loc2 = (28.394231, 77.050308)
hs.haversine(loc1, loc2, unit = Unit.MILES)
```

```python
# getting manhattan distance between start & end for every ride recorded in data

def manhattandistance(row):
    """
    calculate manhattan distance with lat & lng coordinates 
    & convert to miles
    
    """
    sla = row['start_lat']
    slo = row['start_lng']
    ela = row['end_lat']
    elo = row['end_lng']
    d1 = hs.haversine((sla, slo), (ela, slo), unit = Unit.MILES)
    d2 = hs.haversine((ela, slo), (ela, elo), unit = Unit.MILES)
    return d1 + d2
```

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
geolocator = Nominatim(user_agent = 'sample')
geocode = RateLimiter(geolocator.reverse, min_delay_seconds = 0.1)
location = geolocator.reverse("40.709697, -73.907856")

print(location.raw['address'])
```

```python
#create an empty list
#create a sublist and fill the sublist with info
#append the empty list with sublist
#transform that bitch into a dataframe later

startdata = []
enddata = []
key1 = 'neighbourhood'
key2 = 'suburb'
key3 = 'county'

for index, row in subdf.iterrows():

    location = geolocator.reverse("{}, {}".format(row['start_lat'], row['start_lng'])).raw['address']
    mini = []
    mini.append(row['start_lat'])
    mini.append(row['start_lng'])
    if key1 in location:
        mini.append(location['neighbourhood'])
    else:
        mini.append(None)
    if key2 in location:
        mini.append(location['suburb'])
    else:
        mini.append(None)
    if key3 in location:
        mini.append(location['county'])
    else:
        mini.append(None)
    startdata.append(mini)
    print(index, end = '\r')
    

for index, row in enddf.iterrows():

    location = geolocator.reverse("{}, {}".format(row['end_lat'], row['end_lng'])).raw['address']
    mini = []
    mini.append(row['end_lat'])
    mini.append(row['end_lng'])
    if key1 in location:
        mini.append(location['neighbourhood'])
    else:
        mini.append(None)
    if key2 in location:
        mini.append(location['suburb'])
    else:
        mini.append(None)
    if key3 in location:
        mini.append(location['county'])
    else:
        mini.append(None)
    enddata.append(mini)
    print(index, end = '\r')
    
```

```python
start = pd.DataFrame(startdata)

start = start.rename(columns={0: 'start_lat', 
                            1: 'start_lng',
                            2: 'start_hood',
                            3: 'start_boro',
                            4: 'start_county'})

end = pd.DataFrame(enddata)

end = end.rename(columns={0: 'end_lat', 
                            1: 'end_lng',
                            2: 'end_hood',
                            3: 'end_boro',
                            4: 'end_county'})
```
