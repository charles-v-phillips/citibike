```python
os.walk('monthlystation')

# iterating through every csv file in every directory
# reading the csv as a dataframe and exporting it back out as a parquet file

for a,b,c in sorted(os.walk('monthlystation'))[1:]:
        
    p = f'/Users/junglim/Desktop/capstone/citibike/jung/data/stationavailability/{a}/{c[0]}'

    csv1 = pd.read_csv(p, delimiter = '\t', error_bad_lines=False)
    try:
        csv1.to_parquet(p[:-3] + 'parquet')
    except:
        try:
            csv1 = csv1[~csv1['dock_name'].isna()]
            csv1 = csv1[~csv1['avail_bikes'].isna()]
            csv1['dock_id'] = csv1['dock_id'].astype(int)
            csv1['avail_bikes'] = csv1['avail_bikes'].astype(int)
            csv1['_long'] = csv1['_long'].astype(float)
            csv1.to_parquet(p[:-3] + 'parquet')
        except ValueError:
            print('damn')
            
# so it turns out that may, oct, & nov 2015 were unable to be parqueted
# because of more data type errors
    
        
    q = f'/Users/junglim/Desktop/capstone/citibike/jung/data/stationavailability/{a}/{c[1]}'

    csv2 = pd.read_csv(q)
    csv2.to_parquet(q[:-3]+'parquet')
        
        
```

```python
# manually changing csv to parquet in may 2015
fo = pd.read_csv('./monthlystation/2015-05/bikeshare_nyc_raw.csv', delimiter = '\t')
fo = fo[~fo['dock_name'].isna()]
fo = fo[~fo['avail_bikes'].isna()]
fo['dock_id'] = fo['dock_id'].astype(int)
fo['avail_bikes'] = fo['avail_bikes'].astype(int)
fo['avail_docks'] = fo['avail_docks'].astype(int)
fo['status_key'] = fo['status_key'].astype(int)

fo = fo.loc[~fo._long.str.contains("--", na=False)]
fo['_long'] = fo['_long'].astype(float)
fo.to_parquet('./monthlystation/2015-05/bikeshare_nyc_raw.parquet')
```

```python
# manually changing csv to parquet in oct 2015
fe = pd.read_csv('./monthlystation/2015-10/bikeshare_nyc_raw.csv', delimiter = '\t', error_bad_lines = False)
fe = fe.drop(['avail_bikes3', 'dock_name6'], axis = 1)
fe = fe[~fe['dock_name'].isna()]
fe = fe[~fe['avail_bikes'].isna()]
fe['dock_id'] = fe['dock_id'].astype(int)
fe = fe.loc[fe['avail_bikes'] != '33stAddress1\\"\t2\t35\t40.785\t-73.972834\t1\t1\t"\t\\"W 84 St & Columbus Ave\\""']
fe['avail_bikes'] = fe['avail_bikes'].astype(int)

fo.to_parquet('./monthlystation/2015-10/bikeshare_nyc_raw.parquet')
```
