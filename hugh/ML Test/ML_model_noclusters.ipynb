```python
import pandas as pd
import numpy as np
import pickle
```

```python
data = pd.read_csv('ML_table_with_availability_clusters.csv')
data
```

```python
x = data[['dock_id','month', 'num_day', 'hour', 'latitude', 'longitude', 'tot_docks', 
          'tmp', 'rain', 'weekday_cluster', 'weekend_cluster']]
x
```

```python
y = data['avail_bikes']
y
```

```python
no_clusters = pickle.load(open('../cluster_alt_df.pkl', 'rb'))
no_clusters
```

```python
distances = no_clusters[['dock_id', 'cbd_distance', 'subway_distance', 'bus_distance']]
distances
```

```python
df = pd.merge(x, distances, how = 'left', on = 'dock_id')
df
```

```python
df.drop(columns = 'dock_id', inplace = True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn import tree
from sklearn import ensemble
```

```python
np.random.seed(0)
x_train, x_test, y_train, y_test = train_test_split(df, y)
```

```python
rf_model = ensemble.RandomForestRegressor()
```

```python
params = [{
    "max_depth": [30],
    "n_estimators": [500],
    "max_samples": [0.1],
    #"min_samples_leaf": range(1, 10),
    "random_state": [0]}]
grid_search = GridSearchCV(rf_model,params, cv = 2, n_jobs=-1)
grid_search.fit(x_train, y_train)

print('The best parametere are: ', grid_search.best_params_)
print('Train Score: ', grid_search.score(x_train, y_train))
print('Test Score: ', grid_search.score(x_test, y_test))
```
