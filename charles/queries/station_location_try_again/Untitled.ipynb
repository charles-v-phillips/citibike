```python
import pandas as pd
import numpy as np
import geopy
from geopy.geocoders import Nominatim, GoogleV3
from geopy.point import Point
from geopy.extra.rate_limiter import RateLimiter
```

```python
data = pd.read_csv('./hughs_dock_locations.csv')
```

```python
data.drop('occurance', inplace = True,axis = 1)
```

```python
geolocator = Nominatim(user_agent = 'fuck_off')
rv = {}

for i, row in data.iterrows():
    print(f'{i} of {len(data)}',end = '\r')
    addr = geolocator.reverse('{}, {}'.format(row.start_lat,row.start_lon)).raw['address']
    rv[row['start_station_id']] = addr

rv = pd.DataFrame.from_dict(rv)
```

```python
info = rv.T
```

```python
info.reset_index(inplace = True)
```

```python
info = info[['index','neighbourhood','suburb','postcode']]
```

```python
info.rename(columns = {'index': 'dock_id','suburb' : 'borough','postcode': 'zip'},inplace = True)
```
