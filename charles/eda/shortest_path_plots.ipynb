# Using OSRM to get shortest path from point A to point B
OSRM is a service thgat we can use to get the shortest path between 2 points. 


```python
url = 'http://router.project-osrm.org/'
route = f'route/v1/bike/{start_lon},{start_lat};{end_lon},{end_lat}'
request = requests.get(url + route)
```

```python
info = request.json()
```

# Getting the path from the info
From the info, we can get the routes field. I subscript 0 since you may have more than one route returned , although OSRM documentation says the first route in the list os the theoretical fastest. I then get the geometry field, which is a string representation of the path, and decode it

```python
coords = polyline.decode(info['routes'][0]['geometry'])
```
